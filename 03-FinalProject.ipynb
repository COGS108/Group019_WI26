## Abstract

The Summer Olympic Games are a prestigious sporting event held every 4 years since their start in 1896. With the Winter Olympic Games being held in Italy this past February, we were inspired to explore the possibility of predicting Olympic performance by country. As our group holds a passion for public health, we focused this project on discovering a relationship between a country's metrics on several health indices and their subsequent performance in the Olympic Games. Overall, we hypothesized that nations with higher national health metrics perform better in the Olympic Games. Based on historical trends and the development of health standards, we can build a model to predict which nations will win more medals and improve performance over time. 

To conduct this project, we handled two datasets with data ranging from 1976 to 2016. Our first dataset was from the World Bank and included our health and economic indices we used for each participating country. The second dataset used was our Olympic Medals dataset, which contained medal count per country per Summer Olympics. After wrangling and cleaning our datasets, we decided to use forward chaining cross-validation as our chosen model. This is because this model creates a generalizable predictive model while simultaneously avoiding overfitting. 

The first model we created regressed predicted medals based on a country's economic, demographic, and health indices. Our first model achieved an R^2 of ~0.65 and an RMSE of ~9.18. The second model we created removed health indicators and regressed the predicted medal count based on economic and demographic indicators only. This model achieved an R^2 of ~0.617 and an RMSE of ~9.54. Through these scores, we conclude that the inclusion of national health indicators increases the performance of our model on predicting a country's Summer Olympic success. Additional regression analysis examining individual health indicators found that under-5 mortality had a statistically significant negative relationship with medal totals (p < 0.01), while life expectancy and physicians per 1,000 people were not statistically significant once economic and demographic variables were controlled for. These results suggest that while population health indicators provide modest predictive improvements, historical Olympic performance and country size remain the strongest determinants of Olympic medal success.

### Predictive Model with Health, Economic, and Demographic Indicators

In this section, we implement a forward-chaining cross-validation model to predict Olympic medal totals using a combination of health, economic, and demographic indicators. Because Olympic data is time-ordered, forward-chaining cross-validation ensures that models are trained only on past Olympic cycles when predicting future outcomes, preventing information leakage. The model includes variables such as under-5 mortality, life expectancy, physicians per 1,000 people, GDP per capita growth, population, urban population percentage, and previous Olympic medal counts. Forward-chaining cross-validation is used to ensure that models are trained only on past Olympic cycles when predicting future outcomes, preventing information leakage and improving the model’s ability to generalize.

```python
# Model 1: Regressing predicted medals on economic, demographic, and health indicators

import pandas as pd
import numpy as np
import os
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.linear_model import PoissonRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

df = pd.read_csv("final_olympics_dataset.csv").copy()

df = df.sort_values(["Country_code", "year"]).reset_index(drop=True)
df["prev_medals"] = df.groupby("Country_code")["total_medals"].shift(1)
df["prev_medals"] = df["prev_medals"].fillna(0)

target_col = "total_medals"
year_col = "year"

feature_cols = [
    "Country_code",
    "gdp_per_capita_growth",
    "under5_mortality",
    "physicians_per_1000",
    "life_expec_under5",
    "log_population",
    "urban_population_pct",
    "prev_medals"
]

numeric_features = [
    "gdp_per_capita_growth",
    "under5_mortality",
    "physicians_per_1000",
    "life_expec_under5",
    "log_population",
    "urban_population_pct",
    "prev_medals"
]

categorical_features = ["Country_code"]

preprocessor = ColumnTransformer([
    (
        "num",
        Pipeline([
            ("imputer", SimpleImputer(strategy="median")),
            ("scaler", StandardScaler())
        ]),
        numeric_features
    ),
    (
        "cat",
        Pipeline([
            ("imputer", SimpleImputer(strategy="most_frequent")),
            ("onehot", OneHotEncoder(handle_unknown="ignore"))
        ]),
        categorical_features
    )
])

model = Pipeline([
    ("preprocessor", preprocessor),
    ("regressor", PoissonRegressor(alpha=1.0, max_iter=1000))
])

test_years = [1996, 2000, 2004, 2008, 2012, 2016]

all_metrics = []
all_predictions = []

for test_year in test_years:
    train_df = df[df[year_col] < test_year].copy()
    test_df = df[df[year_col] == test_year].copy()

    X_train = train_df[feature_cols]
    y_train = train_df[target_col]
    X_test = test_df[feature_cols]
    y_test = test_df[target_col]

    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    preds = pd.Series(preds, index=test_df.index).clip(lower=0)

    fold_results = test_df[["Country", "Country_code", "year", "total_medals"]].copy()
    fold_results["predicted_medals"] = preds.values
    fold_results["abs_error"] = (fold_results["total_medals"] - fold_results["predicted_medals"]).abs()
    all_predictions.append(fold_results)

    all_metrics.append({
        "test_year": test_year,
        "n_train": len(train_df),
        "n_test": len(test_df),
        "mae": mean_absolute_error(y_test, preds),
        "rmse": mean_squared_error(y_test, preds) ** 0.5,
        "r2": r2_score(y_test, preds)
    })

metrics_df = pd.DataFrame(all_metrics)
predictions_df = pd.concat(all_predictions, ignore_index=True)

# Save for cross-model comparison later in the notebook.
model1_metrics_df = metrics_df.copy()

print(metrics_df)
print()
print(metrics_df[["mae", "rmse", "r2"]].mean().rename("average_across_test_years"))
print()
print(predictions_df.sort_values(["year", "predicted_medals"], ascending=[True, False]))
```

### Baseline Predictive Model without Health Indicators

To evaluate whether health indicators improve predictive performance, we construct a second forward-chaining cross-validation model that includes only economic and demographic variables. This baseline model removes all health indicators while retaining variables such as GDP per capita growth, population, urban population percentage, and previous Olympic medal counts. Comparing this model’s prediction accuracy to the full model allows us to assess whether the inclusion of national health metrics meaningfully improves Olympic medal predictions.

```python
# Model 2: Regressing predicted medals on economic and demographic indicators only (removing health indicators)

import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.linear_model import PoissonRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

df = pd.read_csv("final_olympics_dataset.csv").copy()

df = df.sort_values(["Country_code","year"]).reset_index(drop=True)
df["prev_medals"] = df.groupby("Country_code")["total_medals"].shift(1)
df["prev_medals"] = df["prev_medals"].fillna(0)

target_col = "total_medals"
year_col = "year"

feature_cols = [
    "Country_code",
    "gdp_per_capita_growth",
    "log_population",
    "urban_population_pct",
    "prev_medals"
]

numeric_features = [
    "gdp_per_capita_growth",
    "log_population",
    "urban_population_pct",
    "prev_medals"
]

categorical_features = ["Country_code"]

preprocessor = ColumnTransformer([
    (
        "num",
        Pipeline([
            ("imputer", SimpleImputer(strategy="median")),
            ("scaler", StandardScaler())
        ]),
        numeric_features
    ),
    (
        "cat",
        Pipeline([
            ("imputer", SimpleImputer(strategy="most_frequent")),
            ("onehot", OneHotEncoder(handle_unknown="ignore"))
        ]),
        categorical_features
    )
])

model = Pipeline([
    ("preprocessor", preprocessor),
    ("regressor", PoissonRegressor(alpha=1.0, max_iter=1000))
])

test_years = [1996, 2000, 2004, 2008, 2012, 2016]

all_metrics = []
all_predictions = []

for test_year in test_years:

    train_df = df[df[year_col] < test_year]
    test_df = df[df[year_col] == test_year]

    X_train = train_df[feature_cols]
    y_train = train_df[target_col]

    X_test = test_df[feature_cols]
    y_test = test_df[target_col]

    model.fit(X_train, y_train)

    preds = model.predict(X_test)
    preds = pd.Series(preds, index=test_df.index).clip(lower=0)

    fold_results = test_df[["Country","Country_code","year","total_medals"]].copy()
    fold_results["predicted_medals"] = preds.values
    fold_results["abs_error"] = abs(fold_results["total_medals"] - fold_results["predicted_medals"])

    all_predictions.append(fold_results)

    all_metrics.append({
        "test_year": test_year,
        "mae": mean_absolute_error(y_test, preds),
        "rmse": mean_squared_error(y_test, preds) ** 0.5,
        "r2": r2_score(y_test, preds)
    })

metrics_df = pd.DataFrame(all_metrics)
predictions_df = pd.concat(all_predictions)

# Save for cross-model comparison later in the notebook.
model2_metrics_df = metrics_df.copy()

print(metrics_df)
print()
print(metrics_df[["mae","rmse","r2"]].mean())
```

### Fold-Level Performance Comparison Across Olympic Cycles

In this section, we compare the predictive performance of both forward-chaining models across individual Olympic cycles. Each fold corresponds to a specific Olympic year in which the model is trained on all prior years and evaluated on that year’s medal outcomes. By examining fold-level metrics such as MAE, RMSE, and R^2, we assess how model performance varies across different Olympic cycles and determine whether the inclusion of health indicators consistently improves prediction accuracy over time.

```python
# Fold-level comparison table for Model 1 vs Model 2

# Label model sources and keep a common column layout.
m1 = model1_metrics_df[["test_year", "mae", "rmse", "r2"]].copy()
m1["model"] = "Model 1 (with health)"

m2 = model2_metrics_df[["test_year", "mae", "rmse", "r2"]].copy()
m2["model"] = "Model 2 (no health)"

comparison_long = pd.concat([m1, m2], ignore_index=True)

comparison_table = comparison_long.pivot_table(
    index="test_year",
    columns="model",
    values=["mae", "rmse", "r2"]
)

# Make columns easy to read in notebook output.
comparison_table.columns = [f"{metric} | {model}" for metric, model in comparison_table.columns]
comparison_table = comparison_table.sort_index()

print("Fold-level metric comparison:")
display(comparison_table.round(4))

avg_by_model = comparison_long.groupby("model")[["mae", "rmse", "r2"]].mean()
print("Average metrics across folds:")
display(avg_by_model.round(4))

better_m1 = 0
better_m2 = 0

if avg_by_model.loc["Model 1 (with health)", "mae"] < avg_by_model.loc["Model 2 (no health)", "mae"]:
    better_m1 += 1
else:
    better_m2 += 1

if avg_by_model.loc["Model 1 (with health)", "rmse"] < avg_by_model.loc["Model 2 (no health)", "rmse"]:
    better_m1 += 1
else:
    better_m2 += 1

if avg_by_model.loc["Model 1 (with health)", "r2"] > avg_by_model.loc["Model 2 (no health)", "r2"]:
    better_m1 += 1
else:
    better_m2 += 1

if better_m1 > better_m2:
    print("Conclusion: Model 1 is more optimal overall (wins on more of MAE, RMSE, and R^2).")
elif better_m2 > better_m1:
    print("Conclusion: Model 2 is more optimal overall (wins on more of MAE, RMSE, and R^2).")
else:
    print("Conclusion: Mixed results; each model is better on different metrics.")
```

### Regression Analysis of Health Indicators

To further investigate the relationship between population health and Olympic performance, we conduct a series of regression analyses examining each health indicator individually. Separate ordinary least squares regressions are estimated for under-5 mortality, life expectancy, and physicians per 1,000 people while controlling for economic and demographic variables. Heteroskedasticity-robust (HC3) standard errors are used to obtain reliable statistical inference. These models allow us to evaluate whether specific health indicators are significantly associated with Olympic medal outcomes beyond their contribution to predictive performance.

Now to analyze whether better health metrics (higher life expectancy, higher physicians per capita, lower mortality under five years old) is correlated with a higher number of medals won at the olympics, per our hypothesis.

Our regression model (panel data):

Medals = α + βLife_Expectancy + γ_Under_5_Mortality + δPhysicians_Per_1000 + εGDP_Growth + ζlog(Population) + ηUrban_Population_Percent + θPrevious_Medals + u,

where α is the intercept, the following greek letters are coefficients, and u is the generalized error term.

Because of collinearity between our health variables, we will examine them in three separate regressions, then interpret the results.

```python
import pandas as pd
import statsmodels.api as sm

features = [
    "life_expec_under5",
    "gdp_per_capita_growth",
    "urban_population_pct",
    "log_population",
    "prev_medals"
]

X = df[features]
y = df["total_medals"]

X = sm.add_constant(X)

model = sm.OLS(y, X).fit(cov_type="HC3")

print("OLS regression with health variable: life_expec_under5")
print(model.summary())
```

The only significant (5%) variables were log_population with a positive coefficient, and prev_medals with a positive coefficient. Life_expec_under5 did not have a statistically significant effect, so life expectancy does not appear to be associated with Olympic medal totals once economic and demographic variables are controlled for.

```python
import pandas as pd
import statsmodels.api as sm

features = [
    "under5_mortality",
    "gdp_per_capita_growth",
    "urban_population_pct",
    "log_population",
    "prev_medals"
]

X = df[features]
y = df["total_medals"]

X = sm.add_constant(X)

model = sm.OLS(y, X).fit(cov_type="HC3")

print("OLS regression with health variable: under5_mortality")
print(model.summary())
```

The only significant (5%) variables were log_population with a positive coefficient, prev_medals with a positive coefficient, and under5_mortality with a negative coefficient. Since the coefficient on under5_mortality is negative and significant, this tells us that countries with higher child mortality tend to win fewer Olympic medals.

```python
import pandas as pd
import statsmodels.api as sm

features = [
    "physicians_per_1000",
    "gdp_per_capita_growth",
    "urban_population_pct",
    "log_population",
    "prev_medals"
]

X = df[features]
y = df["total_medals"]

X = sm.add_constant(X)

model = sm.OLS(y, X).fit(cov_type="HC3")

print("OLS regression with health variable: physicians_per_1000")
print(model.summary())
```

The only significant (5%) variables were log_population with a positive coefficient, prev_medals with a positive coefficient, and urban_population_pct with a positive coefficient. Since the coefficient on physicians_per_1000 is insignificant, this tells us that physician density does not appear to be associated with Olympic medal totals once other factors are controlled for.

Regression results using heteroskedasticity-robust (HC3) standard errors show that previous Olympic performance and country population size are the strongest predictors of medal totals. Among the health indicators examined, under-5 mortality is the only variable with a statistically significant relationship with Olympic performance, suggesting that poorer population health conditions may be associated with weaker Olympic outcomes. However, the magnitude of this relationship is small compared to the effects of historical performance and population size.
